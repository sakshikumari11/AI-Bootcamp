```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Sak.csv')

# Plotting
plt.figure(figsize=(14, 8))
plt.bar(df['States and Union Territories'], df['No. Of Phases'], color='skyblue')

# Labels and title
plt.xlabel('States and Union Territories')
plt.ylabel('No. Of Phases')
plt.title('Number of Phases in Different States and UTs')
plt.xticks(rotation=90)  # Rotate state names for readability
plt.tight_layout()

# Display the chart
plt.show()
```
